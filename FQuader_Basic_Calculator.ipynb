```python
import math
```

```python
def multiply(firstNum, secondNum):
    product = 0
    absolute_fNum = abs(firstNum)
    absolute_sNum = abs(secondNum)
    if firstNum == 0 or secondNum == 0:
        return 0
    for x in range(0, absolute_sNum):           #Arithmetic patterns are looped to calculate for product
        product += absolute_fNum
    if firstNum < 0 or secondNum < 0:           #Numbers are checked for negative and positive values
        if firstNum < 0 and secondNum < 0:
            return product
        return '-' + str(product)
    else:
        return product
```

```python
def divide_smaller(dividend):
    return multiply(dividend, 10)
```

```python
def divide_greater(dividend, divisor, absolute_divid, absolute_divis):
    resultant = 0
    full_resultant = 0
    timer = 0
    while (True):
        while(absolute_divid != 0):                                                                          #Process checks for whole numbers and decimal numbers
            absolute_divid -= absolute_divis
            if (absolute_divid < absolute_divis):                                                            #Process checks if remainder is less than dividend
                absolute_divid = divide_smaller(absolute_divid)
                timer += 1
                if timer == 1:
                    resultant_append = resultant + 1
                    resultant = 0
                elif timer == 2:
                    resultant_append2 = resultant
                    resultant = 0
                elif timer == 3:                                                                             #Process stops after the second decimal place has been given
                    resultant_append3 = resultant
                    return str(resultant_append) + "." + str(resultant_append2) + str(resultant_append3)            
            if timer == 1:
                full_resultant = str(resultant_append) + "." + str(resultant)
            elif timer == 2:
                full_resultant = str(resultant_append) + "." + str(resultant_append2) + str(resultant)
            resultant += 1
        if dividend < 0:                                                                                     #Process checks for negative and positive numbers
            if divisor < 0:
                return full_resultant
            return '-' + str(full_resultant)
        elif divisor < 0:
            if dividend < 0:
                return full_resultant
            return '-' + str(full_resultant)
        else:
            return full_resultant
```

```python
def prime_check(num): 
    if num == 1:                                                  #Checks for obvious non-prime numbers before continuing
        return False  
    for x in range(2, (int)(math.sqrt(num))+1):                   #The loop ends at the included square root of the number
        if num % x == 0:                                          #If the number has a remainder of zero in a certain iteration, it proceeds the num is not prime
            return False 
    return True
```
